```python
def dept_no_train(x):
    if x['department']=='Legal':
        if x['no_of_trainings']>=legal:
            return 1
        else:
            return 0
    elif x['department']=='HR':
        if x['no_of_trainings']>=hr:
            return 1
        else:
            return 0
    elif x['department']=='Finance':
        if x['no_of_trainings']>=fina:
            return 1
        else:
            return 0
    elif x['department']=='Procurement':
        if x['no_of_trainings']>=proc:
            return 1
        else:
            return 0
    elif x['department']=='R&D':
        if x['no_of_trainings']>=r_d:
            return 1
        else:
            return 0
    elif x['department']=='Analytics':
        if x['no_of_trainings']>=analy:
            return 1
        else:
            return 0
    elif x['department']=='Technology':
        if x['no_of_trainings']>=tech:
            return 1
        else:
            return 0
    elif x['department']=='Operations':
        if x['no_of_trainings']>=op:
            return 1
        else:
            return 0
    elif x['department']=='Sales & Marketing':
        if x['no_of_trainings']>=s_and_m:
            return 1
        else:
            return 0
```
