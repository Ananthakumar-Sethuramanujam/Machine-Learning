# Lowercase abstract and article title

# Lemmatize

```python
from textblob import Word
info_train['abstract'] = info_train['abstract'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
info_train['article_title']=info_train['article_title'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```

# Map the words to (word, 1) format

```python
import collections
def mapred(x):
    wordlist=x.split(' ')
    l = [(x,1) for x in wordlist]
    c = collections.Counter()
    for a,i in l:
        c[a] += i
    return c
info_train['abstract']=info_train['abstract'].apply(mapred)
```

```python
info_train['article_title']=info_train['article_title'].apply(mapred)
```

# Convert publication date to date since publication

```python
import datetime

datetime.date.today()
```

```python
import datetime
from datetime import datetime
def find_days(x):
    start = datetime.strptime(x, '%Y-%m-%d').date()
    today = datetime.now().strftime("%Y-%m-%d")
    today = datetime.strptime(today, '%Y-%m-%d').date()
    return (today-start).days
info_train['days_since_publication']=info_train['pub_date'].apply(find_days)
```

# Sort predictor_kw based on value and take top 10

```python
def take_first_ten(x):
    check=dict(x)
    check=sorted(check.items(), key=lambda x: x[1],reverse=True)
    check=list(check)
    return check[0:10]
train['predictor_kw']=train['predictor_kw'].apply(take_first_ten)
```

```python
train.head()
```

# Find the solution

For each article, find the number of common words with all other articles that are in the same set and published earlier

```python
def find_results1(x):
    res_list=[]
    d=dict(x['predictor_kw'])
    dict1=list(d.keys())
    df=train[(train['set']==x['set']) & (train['date']>x['date'])] # compare with articles that are published before the current one
    for i in df.iterrows():
        ids=i[1]['pmid']
        lists=list(dict(i[1]['predictor_kw']).keys())
        count=len(list(set(dict1).intersection(lists)))
        res_list.append((ids,count))
    return res_list
train['result_count']=train.apply(find_results1,axis=1)
```

```python
def sort_return(x):
    check=dict(x)
    check=sorted(check.items(), key=lambda x: x[1],reverse=True)
    check=list(check)
    return check[0:13]
train['result_count']=train['result_count'].apply(sort_return)
```

```python
train.head()
```

```python
def return_only_keys(x):
    r=dict(x)
    return list(r.keys())
train['result_count']=train['result_count'].apply(return_only_keys)
```
