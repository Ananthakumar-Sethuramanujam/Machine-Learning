```python
import pandas as pd
import numpy as np
```

# Data Preparation

```python
data=pd.read_csv('data.csv')
```

```python
data.shape
```

```python
pd.set_option('display.max_columns',None)
data.head()
```

```python
data.isnull().sum()
```

```python
labels=data.y
data1=data.iloc[:,1:data.shape[1]-1]
data1.shape
```

```python
data1.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data1)
data1=scaler.transform(data1)
data1=pd.DataFrame(data1)
data1.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
def labels_code(x):
    if x==1:
        return 1
    else:
        return 0
```

```python
labels=labels.apply(labels_code)
```

```python
labels.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
     data1, labels, test_size=0.30, random_state=42)
```

```python
x_train.shape
```

# Model Building

```python
from keras.models import Sequential
from keras.layers import Dense, Activation,Dropout
from keras.optimizers import SGD


model = Sequential()

model.add(Dense(units=128, activation='softmax', input_dim=178))
model.add(Dropout(0.5))
model.add(Dense(units=128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=128, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=2, activation='softmax'))

opt = SGD(lr=0.01)
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
y_train=pd.get_dummies(y_train)
y_test=pd.get_dummies(y_test)
```

```python
y_train.apply(lambda x:sum(x)/y_train.shape[0])
```

```python
y_test.apply(lambda x:sum(x)/y_test.shape[0])
```

```python
model.fit(x_train, y_train,
          batch_size=64, epochs=50,
          validation_data=(x_test, y_test))
```

```python
pr=model.predict(x_test)
```

```python
pr
```

```python
y_test_argmax=np.argmax(np.array(y_test),axis=1)
```

```python
y_test_argmax
```

```python
pr_argmax=np.argmax(pr,axis=1)
pr_argmax
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test_argmax,pr_argmax))
```

# Balancing the Data and Retraining the model

```python
x_train.head()
```

```python
y_train.head()
```

```python
y_train1=pd.DataFrame(np.argmax(np.array(y_train),axis=1))
y_train1.head()
```

```python
x_train.reset_index(inplace=True,drop=True)
```

```python
full=pd.concat([x_train,y_train1],axis=1)
full.head()
```

```python
onlyzero=full[full.iloc[:,-1]==0]
```

```python
onlyone=full[full.iloc[:,-1]==1]
```

```python
onlyzero.shape,onlyone.shape
```

```python
onlyz1=onlyzero.sample(frac=0.4)
onlyz1.shape
```

```python
finale=pd.concat([onlyz1,onlyone])
```

```python
finale.shape
```

```python
df = finale.sample(frac=1).reset_index(drop=True)
```

```python
df.head()
```

```python
labels_balanced=df.iloc[:,-1]
```

```python
labels_balanced=pd.get_dummies(labels_balanced)
labels_balanced.head()
```

```python
df1=df.iloc[:,:-1]
df1.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation,Dropout
from keras.optimizers import SGD


model = Sequential()

model.add(Dense(units=128, activation='softmax', input_dim=178))
model.add(Dropout(0.5))
model.add(Dense(units=128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=128, activation='softmax'))
model.add(Dropout(0.5))
model.add(Dense(units=128, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=2, activation='softmax'))

opt = SGD(lr=0.01)
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(df1, labels_balanced,
          batch_size=64, epochs=30,
          validation_data=(x_test, y_test)) # validation data won't be used for training
```

```python
labels_balanced1=np.argmax(np.array(labels_balanced),axis=1)
```

```python
pp=model.predict(x_test)  # predict on test data
pp1=np.argmax(pp,axis=1)
```

```python
print(classification_report(y_test_argmax,pp1))
```
